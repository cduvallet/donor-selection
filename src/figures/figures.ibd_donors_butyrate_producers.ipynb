This notebook is my first attempt to make "final" (i.e. prettier) figures for the IBD case study.

```python
import pandas as pd
import numpy as np

from scipy.stats import ttest_ind

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
%matplotlib inline

sns.set_palette('muted')
```

This plot will have a top row that's the abundance of butyrate producing bacteria in each donor sample for the three studies, and a bottom row that has the abundance of butyrate producers split by yes/no FMT response.

Also for the studies which measured FMT response at different times or in different ways, we'll also make all of those figures here.

# Main figure

```python
# File names for donor butyrate abundance
fjacob = '../../data/analysis/butyrate_producers.jacob2017.txt'
fgoyal = '../../data/analysis/butyrate_producers.goyal2018.txt'
fkump = '../../data/analysis/butyrate_producers.kump2018.txt'

donorbut_jacob = pd.read_csv(fjacob, sep='\t')
donorbut_jacob = donorbut_jacob.query('donor_patient == "donor"')

donorbut_goyal = pd.read_csv(fgoyal, sep='\t')
donorbut_goyal = donorbut_goyal.query('time_point == "donor"')

donorbut_kump = pd.read_csv(fkump, sep='\t')
donorbut_kump = donorbut_kump.query('Sampling_day == "average"')
donorbut_kump = donorbut_kump.query('Response != "PR"')
```

```python
plt.style.use('seaborn-white')

total_width = 10
height = 2

xlabel = 'Donor'
ylabel = 'Abundance'

# Set up gridspec
n_jacob = int(donorbut_jacob.shape[0])
n_goyal = int(donorbut_goyal.shape[0])
n_kump = int(donorbut_kump.shape[0])

total_length = n_jacob + n_goyal + n_kump

# In matplotlib 2.2.2, the following code without the additional subplot slots in
# between each gridspec works with a call to gs.tight_layout(fig)
# at the end. Howver, in 2.2.3 this no longer works. Instead, I'm adding some
# empty slots in between each plot (i.e. when calling plt.subplot(gs[:]), I'm
# starting each new subplot 3 after the end of the previous one).
# Note: "no longer works" means that I get the following error: "UserWarning: 
# tight_layout cannot make axes width small enough to accommodate all axes decorations"

## Set up figure
fig = plt.figure(figsize=(total_width, height))
gs = gridspec.GridSpec(nrows=1, ncols=n_jacob + n_goyal + n_kump + 6, 
                       wspace=0.25, figure=fig)

## Jacob
ax = plt.subplot(gs[:n_jacob])

ax.bar(range(n_jacob), 
       donorbut_jacob.sort_values(by='butyrate_abun', ascending=False)['butyrate_abun'].values,
       color='gray')

ax.set_xlabel(xlabel, fontsize='x-large')
ax.set_xlim([-0.5, n_jacob - 0.5])
ax.set_xticks([float(i) + 0.5 for i in range(n_jacob)])
ax.set_xticklabels('')

ax.set_ylabel(ylabel, fontsize='x-large')
ax.set_yticks(ax.get_yticks()[::2])
ax.set_title('Jacob 2017', fontsize='xx-large')

## Goyal
ax = plt.subplot(gs[n_jacob+3:(n_jacob + n_goyal)+3])

ax.bar(range(n_goyal), 
       donorbut_goyal.sort_values(by='butyrate_abun', ascending=False)['butyrate_abun'].values,
       color='gray')

ax.set_xlim([-0.5, n_goyal - 0.5])
ax.set_xticks([float(i) + 0.5 for i in range(n_goyal)])
ax.set_xlabel(xlabel, fontsize='x-large')
ax.set_xticklabels('')

ax.set_ylabel('')
#ax.set_yticks(ax.get_yticks()[::2])
ax.set_yticks([0, 0.1, 0.2, 0.3, 0.4])
ax.set_title('Goyal 2018', fontsize='xx-large')

## Kump
ax = plt.subplot(gs[(n_jacob + n_goyal)+6:])
ax.bar(range(n_kump), 
       donorbut_kump.sort_values(by='butyrate_abun', ascending=False)['butyrate_abun'].values,
       color='gray')

ax.set_xlabel(xlabel, fontsize='x-large')
ax.set_xlim([-0.5, n_kump - 0.5])
ax.set_xticks([float(i) + 0.5 for i in range(n_kump)])
ax.set_xticklabels('')

ax.set_ylabel('')
#ax.set_yticks(ax.get_yticks()[::2])
ax.set_yticks([0, 0.1, 0.2, 0.3, 0.4])
ax.set_title('Kump 2018', fontsize='xx-large')

gs.tight_layout(fig)

fig.savefig('../../figures/final/fig2.donor_butyrate_abun.png', dpi=150)
```

```python
# Bottom panel: abundance of butyrate producers vs. FMT response
total_width = 10
height = 3

fig = plt.figure(figsize=(total_width, height))

gs = gridspec.GridSpec(nrows=1, ncols=3, 
                       wspace=0.5,
                       top=0.55, bottom=0.05)

boxprops = {'edgecolor': 'gray', 'facecolor': 'w', 'linewidth': 0.8}
lineprops = {'color': 'gray', 'linewidth': 0.8}
boxplot_kwargs = dict({'boxprops': boxprops, 'medianprops': lineprops,
                       'whiskerprops': lineprops, 'capprops': lineprops,
                       'width': 0.75})

# Jacob 2017
ax = plt.subplot(gs[0])
sns.boxplot(data=donorbut_jacob, x='response_w4', y='butyrate_abun', ax=ax,
              order=['Yes', 'No'],
            fliersize=0,
            **boxplot_kwargs)
sns.stripplot(data=donorbut_jacob, x='response_w4', y='butyrate_abun', ax=ax,
              order=['Yes', 'No'],
              size=6, linewidth=0.2)
ax.set_title('Jacob 2017', fontsize='large')
ax.set_xlabel('')
ax.set_ylabel('Donor abundance of\nbutyrate producers', fontsize='large')
ax.set_xticklabels(['Response', 'No response'], fontsize='large')

ax.set_ylim([0, 0.4])
#ax.set_yticks(ax.get_yticks()[::2])
ax.set_yticks([0, 0.1, 0.2, 0.3, 0.4])


# Goyal 2018
ax = plt.subplot(gs[1])
sns.boxplot(data=donorbut_goyal, x='response_m6', y='butyrate_abun', ax=ax,
            order=[True, False],
            fliersize=0,
            **boxplot_kwargs)
sns.stripplot(data=donorbut_goyal, x='response_m6', y='butyrate_abun', ax=ax,
              order=[True, False],
              size=6, linewidth=0.2)
ax.set_title('Goyal 2018', fontsize='large')
ax.set_xlabel('')
ax.set_xticklabels(['Response', 'No response'], fontsize='large')

ax.set_ylim([0, 0.4])
ax.set_ylabel('')
#ax.set_yticks(ax.get_yticks()[::2])
ax.set_yticks([0, 0.1, 0.2, 0.3, 0.4])


# Kump 2018
ax = plt.subplot(gs[2])
sns.boxplot(data=donorbut_kump, x='Response', y='butyrate_abun', ax=ax,
            order=['RE', 'NR'],
            fliersize=0,
            **boxplot_kwargs)
sns.stripplot(data=donorbut_kump, x='Response', y='butyrate_abun', ax=ax,
              order=['RE', 'NR'],
              size=6, linewidth=0.2)
ax.set_title('Kump 2018', fontsize='large')
ax.set_xlabel('')
ax.set_xticklabels(['Response', 'No response'], fontsize='large')

ax.set_ylim([0, 0.4])
ax.set_ylabel('')
ax.set_yticks([0, 0.1, 0.2, 0.3, 0.4])

gs.tight_layout(fig)

fig.savefig('../../figures/final/fig2.butyrate_vs_response.png', dpi=150)
```

### P values

```python
# Jacob 2017
#response_col = 'response_w4'
responder = 'Yes'
non_responder = 'No'

x = donorbut_jacob.query('response_w4 == @responder')['butyrate_abun'].values
y = donorbut_jacob.query('response_w4 == @non_responder')['butyrate_abun'].values
print("Jacob: ", ttest_ind(x, y))

# Goyal 2018
#response_col = 'response_m6'
responder = True
non_responder = False

x = donorbut_goyal.query('response_m6 == @responder')['butyrate_abun'].values
y = donorbut_goyal.query('response_m6 == @non_responder')['butyrate_abun'].values
print("Goyal: ", ttest_ind(x, y))


# Kump 2018
#response_col = 'Response'
responder = 'RE'
non_responder = 'NR'

x = donorbut_kump.query('Response == @responder')['butyrate_abun'].values
y = donorbut_kump.query('Response == @non_responder')['butyrate_abun'].values
print("Kump: ", ttest_ind(x, y))
```

### Goyal 2018

```python
def goyal_onecol(ax, col):
    sns.boxplot(data=donorbut_goyal, x=col, y='butyrate_abun', ax=ax,
                order=[True, False],
                fliersize=0,
                **boxplot_kwargs)
    sns.stripplot(data=donorbut_goyal, x=col, y='butyrate_abun', ax=ax,
                  order=[True, False],
                  size=6, linewidth=0.2)
    ax.set_xlabel('')
    ax.set_xticklabels(['Response', 'No response'], fontsize='large')
    ax.set_ylim([0, None])
    ax.set_ylabel('')
```

```python
fig, AX = plt.subplots(2, 2, figsize=(8, 8))
AX = AX.flatten()

ax = AX[0]
col = 'remission_m1'
goyal_onecol(ax, col)
ax.set_title('Remission at M1')

ax = AX[1]
col = 'response_m1'
goyal_onecol(ax, col)
ax.set_title('Response at M1')

ax = AX[2]
col = 'remission_m6'
goyal_onecol(ax, col)
ax.set_title('Remission at M6')

ax = AX[3]
col = 'response_m6'
goyal_onecol(ax, col)
ax.set_title('Response at M6')

fig.tight_layout()

fig.savefig('../../figures/final/suppfig.butyrate_vs_response.goyal2018_all_outcomes.png', dpi=150)
```
