This notebook is an exploration of the OTU-level signal to noise ratio vs. the genus-level, in response to a reviewer's comment that we expect the genus-level simulation to underestimate the power needed (i.e. to have a stronger signal than OTU-level).

I only include analyses for the non-Goodrich datasets, because it took too long to calculate the OTU-level signal to noise.

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
baxter_otu = pd.read_csv('../../data/analysis/population_effects.otu.crc_baxter.txt', sep='\t')
baxter_genus = pd.read_csv('../../data/analysis/population_effects.crc_baxter.txt', sep='\t')
```

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].hist(baxter_otu['snr'])
ax[0, 0].set_title('OTU signal to noise')

ax[0, 1].hist(baxter_genus['snr'])
ax[0, 1].set_title('Genus signal to noise')

ax[1, 0].hist(baxter_otu['abs_snr'])
ax[1, 0].set_title('OTU |signal to noise|')

ax[1, 1].hist(baxter_genus['abs_snr'])
ax[1, 1].set_title('Genus |signal to noise|')

fig.tight_layout()

print('CRC Baxter')
```
