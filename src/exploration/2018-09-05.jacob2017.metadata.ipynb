This notebook cleans up the Jacob 2017 metadata file(s)

```python
import pandas as pd
```

```python
# This file was acquired via email communication, and should not be public
foutcomes = '../../data/raw/jacob2017/FMT Clinical Response Remission.xlsx'
# This file was downloaded from the ENA. 
fmeta = '../../data/raw/jacob2017/jacob2017.PRJNA388210.txt'

fout = '../../data/clean/jacob2017.metadata.txt'
```

```python
outcomes = pd.read_excel(foutcomes)
meta = pd.read_csv(fmeta, sep='\t')
```

```python
# Expand the sample ID to different columns
sampledf = meta['sample_alias'].str.split('.', expand=True)
sampledf.head()
```

```python
# Replace 'Donor' with 'donor'
sampledf = sampledf.replace('Donor', 'donor')
```

Now, let's combine this back with the metadata.

```python
sampledf = sampledf.rename(
    columns={0: 'study_id',
             1: 'FMT', 
             2:'patient_id',
             3: 'sample_id'})
sampledf.head()
```

```python
allmeta = pd.merge(
    sampledf[['patient_id', 'sample_id']], meta, 
    left_index=True, right_index=True)
```

```python
allmeta.head()
```

```python
outcomes
```

Change the patient ID to an integer in allmeta, so that it matches the outcome data (which is an Excel file, so was automatically converted to an int)

```python
allmeta['patient_id'] = allmeta['patient_id'].astype(int)
```

```python
outcomes = outcomes.rename(
    columns={'Remission W4 ': 'remission_w4',
             'Response W4 ': 'response_w4'})
```

```python
allmeta = pd.merge(allmeta, outcomes, left_on='patient_id', right_on='Patient')
allmeta.head()
```

## Add sample type indicator

I need to add a column that indicates whether the sample is a patient or donor sample.

```python
allmeta['sample_type'] = allmeta['sample_id'].apply(lambda x: 'donor' if x == 'donor' else 'patient')
```

## Finalize metadata and write to file

```python
# Reorder columns for my own sanity
first_cols = ['run_accession', 'patient_id', 'sample_id', 'sample_type', 'remission_w4', 'response_w4', 'sample_title']
col_order = first_cols + [i for i in allmeta.columns if i not in first_cols]
allmeta = allmeta[col_order]
allmeta.head()
```

Let's do some checks and make sure I didn't duplicate or lose samples in all the merging

```python
print(sampledf.groupby(['patient_id', 'sample_id']).size().reset_index()[0].unique())

sampledf.groupby(['patient_id', 'sample_id']).size()
```

```python
# Make sure there aren't duplicate samples
print(allmeta.groupby(['patient_id', 'sample_id']).size().reset_index()[0].unique())

# Just look at the data
allmeta.groupby(['patient_id', 'sample_id']).size()
```

Okay, looks good. Let's write this to file!

```python
allmeta.to_csv(fout, sep='\t', index=False)
```
